```python
import cv2

def get_frames(video, save, name_data, id_data):
    indices = []

    cap = cv2.VideoCapture(video)
    video1 = video.split("/")[-1]
    i = 0
    # a variable to set how many frames you want to skip
    frame_skip = cap.get(cv2.CAP_PROP_FPS) - 1
    # a variable to keep track of the frame to be saved
    frame_count = 0
    while cap.isOpened():
      ret, frame = cap.read()
      if not ret:
          break
      if i > frame_skip - 1:
          frame_count += 1
          filename = str(id_data[name_data.index(video1)]) + '_'+str(frame_count)+'sec.jpg'
          cv2.imwrite(filename, frame)
          shutil.copyfile(filename, save + filename)
          print('frame at second ' + str(frame_count) + ' of video ' + str(id_data[name_data.index(video1)]) + ' successfully created')
          indices.append(str(frame_count) + ' ' + str(video))
          i = 0
          continue
      i += 1

    cap.release()
    return indices
```

```python
from ollama import chat
import base64
from PIL import Image
import imageio.v3 as iio
model = 'qwen3-vl:4b'

def run_ollama_chat(img, prompt):
  img_data = iio.imwrite("<bytes>", img, extension='.jpg')


  stream = chat(
      model=model,
      messages=[{'role': 'user', 'content': prompt,
        'images': [base64.b64encode(img_data).decode('utf-8')]}],
      stream=True,
  )
  response = ""
  for chunk in stream:
    response += chunk['message']['content']
  return response
```
