# functions

```python
def get_frames(video, save, name_data, id_data):
    indices = []

    cap = cv2.VideoCapture(video)
    video1 = video.split("/")[-1]
    i = 0
    # a variable to set how many frames you want to skip
    frame_skip = cap.get(cv2.CAP_PROP_FPS) - 1
    # a variable to keep track of the frame to be saved
    frame_count = 0
    while cap.isOpened():
      ret, frame = cap.read()
      if not ret:
          break
      if i > frame_skip - 1:
          frame_count += 1
          filename = str(id_data[name_data.index(video1)]) + '_'+str(frame_count)+'_frame.jpg'
          cv2.imwrite(filename, frame)
          #shutil.copyfile(filename, save + filename)
          print('frame at second ' + str(frame_count) + ' of video ' + str(id_data[name_data.index(video1)]) + ' successfully created')
          indices.append(str(frame_count) + ' ' + str(video))
          i = 0
          continue
      i += 1

    cap.release()
    return indices
```

# split frames

```python
id = []
video_name = []

for x in os.listdir(base + 'videos/bbc'):
  video_name.append(x)
  id.append('bbc' + str(len(id)))

video_id = list(zip(video_name, id))
```

```python
for x in os.listdir(base + '/videos/bbc'):
  get_frames(base + 'videos/bbc/' + x, 'NA', video_name, id)
```
